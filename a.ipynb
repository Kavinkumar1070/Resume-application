```python
import PyPDF2
import re
import spacy

# Load NLP model (spaCy)
nlp = spacy.load("en_core_web_sm")

def extract_text_from_pdf(pdf_path):
    with open(pdf_path, "rb") as file:
        reader = PyPDF2.PdfReader(file)
        text = ""
        for page in reader.pages:
            text += page.extract_text()
    return text

def extract_personal_details(text):
    # Example regex for email and phone
    email = re.findall(r'\S+@\S+', text)
    phone = re.findall(r'\(?\d{3}\)?[-.\s]?\d{3}[-.\s]?\d{4}', text)
    return {
        "email": email[0] if email else None,
        "phone": phone[0] if phone else None
    }

def extract_skills(text):
    skills = []
    # Sample skill list
    skill_keywords = ["Python", "Java", "Data Analysis", "Machine Learning", "SQL"]
    
    for skill in skill_keywords:
        if skill.lower() in text.lower():
            skills.append(skill)
    
    return skills

def extract_experience(text):
    doc = nlp(text)
    experiences = []
    
    for ent in doc.ents:
        if ent.label_ == "ORG":  # Using spaCy's NER to find company names
            experiences.append(ent.text)
    
    return experiences

# Example usage
pdf_path = "resume.pdf"
resume_text = extract_text_from_pdf(pdf_path)
personal_details = extract_personal_details(resume_text)
skills = extract_skills(resume_text)
experience = extract_experience(resume_text)

print("Personal Details:", personal_details)
print("Skills:", skills)
print("Experience:", experience)
```

```python
from docx import Document

def create_resume_doc(personal_details, skills, experience):
    doc = Document()

    # Add personal details
    doc.add_heading(personal_details['name'], 0)
    doc.add_paragraph(f"Email: {personal_details['email']}")
    doc.add_paragraph(f"Phone: {personal_details['phone']}")

    # Add skills section
    doc.add_heading('Skills', level=1)
    doc.add_paragraph(', '.join(skills))

    # Add experience section
    doc.add_heading('Experience', level=1)
    for exp in experience:
        doc.add_paragraph(exp)

    # Save document
    doc.save('Formatted_Resume.docx')

# Example usage
personal_details = {
    "name": "John Doe",
    "email": "johndoe@example.com",
    "phone": "123-456-7890"
}
skills = ["Python", "Data Analysis", "Machine Learning"]
experience = ["Company A", "Company B"]

create_resume_doc(personal_details, skills, experience)
```

1. Extracting Keywords from Job Description

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import re

# Sample job description and resume content
job_description = """
We are looking for a skilled Data Scientist with experience in Python, Machine Learning, and Data Visualization. 
The ideal candidate will have expertise in statistical modeling, data mining, and working with large datasets.
Familiarity with tools like TensorFlow, PyTorch, and SQL is a plus.
"""

resume = """
John Doe is a Data Scientist experienced in Python and data analysis. He has worked with machine learning models and built visualizations using matplotlib.
"""

# Function to clean and preprocess text
def preprocess_text(text):
    text = re.sub(r'[^\w\s]', '', text)  # Remove punctuation
    return text.lower()

# Preprocessing job description and resume
job_description_cleaned = preprocess_text(job_description)
resume_cleaned = preprocess_text(resume)

# Using TF-IDF to extract keywords from the job description
vectorizer = TfidfVectorizer(stop_words='english')
tfidf_matrix = vectorizer.fit_transform([job_description_cleaned])
feature_names = vectorizer.get_feature_names_out()

# Extract top keywords from job description
def get_top_keywords(tfidf_matrix, feature_names, top_n=10):
    sorted_items = tfidf_matrix.toarray().argsort()[:, -top_n:][0][::-1]
    keywords = [feature_names[i] for i in sorted_items]
    return keywords

job_keywords = get_top_keywords(tfidf_matrix, feature_names, top_n=10)

print("Top Keywords from Job Description:", job_keywords)
```

2. Compare Keywords with Resume Content

```python
# Extract keywords from the resume (using simple word matching in this case)
resume_keywords = set(resume_cleaned.split())

# Finding missing keywords in the resume that are in the job description
missing_keywords = [keyword for keyword in job_keywords if keyword not in resume_keywords]

print("Suggested Keywords to Add to Resume:", missing_keywords)
```
